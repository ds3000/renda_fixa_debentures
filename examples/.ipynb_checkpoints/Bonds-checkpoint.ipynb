### Descrição do título

```python
bond.des()
```

Lista de parâmetros da função:

- **convention**: convenção do período de capitalização do juros;
- **frequency**: frequência de pagamentos dos cupons;
- **current_date**: Data na qual deseja calcular os parâmetros do título;
- **maturity_date**: Data de vencimento do título;
- **last_coupon_date**: Data do último pagamento de cupom recebido;
- **next_coupon_date**: Data do pagamento de cupom subsequente;
- **days_to_next_coupon**: Dias úteis até o próximo pagamento de cupom;
- **remaining_payments**: Número remanescente de entradas do fluxo caixa;
- **term_to_maturity_years**: Prazo para o vencimento em anos;
- **coupon_rate**: Taxa do cupom ao ano;
- **coupon_value**: Valor do cupom pago;
- **nominal_value**: Valor nominal ou de face do título;
- **price**: Preço/cotação do título;
- **ytm**: *Yield to maturity* ou TIR do título;
- **current_yield**: *Current yield* (valor do cupom / preço atual);
- **duration**: Duration do título em dias úteis;
- **modified_duration**: Duration modificada em anos;
- **dv01**: *Dollar Duration* - duration em valor monetário para delta_ytm de 1 bps;
- **convexity**: Convexidade do título;
- **sensibility**: Variação percentual do preço do título em resposta à delta_ytm de 100 bps.

Use a função `get()` para obter uma característica específica.

```python
bond.get("price")
```

```python
bond.get("ytm")
```

```python
bond.get("duration")
```

### Bond Curve

```python
series_coupon_bond = bond.bond_curve()
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(x="Date", y="Price", data=series_coupon_bond)
```

```python
series_coupon_bond.tail()
```
